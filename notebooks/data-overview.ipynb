```python
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import image as img

from config import EXTERNAL_DATA_DIR
```

```python
images_dir = EXTERNAL_DATA_DIR / 'images'
assert images_dir.exists()
```

```python
info_df = pd.read_csv(EXTERNAL_DATA_DIR / 'info.csv', index_col=0).reset_index(drop=True)

info_df.head(2)
```

```python
fig, axes = plt.subplots(ncols=3, nrows=3)
fig.set_size_inches(10, 8)

for i in range(3):
    for j, fname in enumerate(info_df.loc[i, ['ImageFile', 'DepthFile', 'SegFile']]):
        fpath = images_dir / fname
        image = img.imread(fpath)
        
        ax = axes[i, j]
        ax.imshow(image)
        ax.axis('off')
```
